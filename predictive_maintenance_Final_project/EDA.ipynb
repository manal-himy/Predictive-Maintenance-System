```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import os
```

```python
# التأكد من وجود مجلد static لحفظ الصور للعرض في الويب
if not os.path.exists('static'):
    os.makedirs('static')
```

```python
df = pd.read_csv('data/predictive_maintenance.csv')
```

```python
# ملاحظة: استبدل 'df' باسم متغير البيانات لديك في الكود
# --- الرسم الأول: علاقة العزم بالسرعة (Scatter Plot) ---
# يوضح كيف ينهار التوازن بين السرعة والعزم ويؤدي للفشل
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Rotational speed [rpm]', y='Torque [Nm]', hue='Failure Type', data=df, alpha=0.5)
plt.title('Relationship: Rotational Speed vs Torque')
plt.xlabel('Speed (rpm)')
plt.ylabel('Torque (Nm)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig('static/1_speed_vs_torque.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# ملاحظة: استبدل 'df' باسم متغير البيانات لديك في الكود
# --- الرسم الأول: علاقة العزم بالسرعة (Scatter Plot) ---
# يوضح كيف ينهار التوازن بين السرعة والعزم ويؤدي للفشل
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Rotational speed [rpm]', y='Torque [Nm]', hue='Failure Type', data=df, alpha=0.5)
plt.title('Relationship: Rotational Speed vs Torque')
plt.xlabel('Speed (rpm)')
plt.ylabel('Torque (Nm)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig('static/1_speed_vs_torque.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# --- الرسم الثاني: تأثير حرارة العملية على الفشل (Regression Plot) ---
# يوضح اتجاه تأثير الحرارة وزيادة احتمالية العطل
plt.figure(figsize=(10, 6))
sns.regplot(x='Process temperature [K]', y='Target', data=df, scatter_kws={'alpha': 0.1}, line_kws={'color': 'red'})
plt.title('Impact of Process Temperature on Failure Probability')
plt.xlabel('Process Temperature (K)')
plt.ylabel('Failure Probability')
plt.savefig('static/2_temp_impact.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# --- الرسم الثالث: توزيع تآكل العدة حسب نوع المنتج (Boxplot) ---
# يوضح أي فئة منتجات (H, M, L) تستهلك العدة أسرع
plt.figure(figsize=(10, 6))
sns.boxplot(x='Type', y='Tool wear [min]', data=df, palette='Set2')
plt.title('Tool Wear Distribution per Product Type')
plt.xlabel('Product Type (L, M, H)')
plt.ylabel('Tool Wear (min)')
plt.savefig('static/3_tool_wear_dist.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python

# --- الرسم الرابع: تحليل كثافة الحرارة (Density Plot) ---
# يوضح تداخل حرارة الهواء مع حرارة العملية عند حدوث الأعطال
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Air temperature [K]', y='Process temperature [K]', cmap="Reds", fill=True)
plt.title('Density Analysis: Thermal States of the Machine')
plt.xlabel('Air Temp (K)')
plt.ylabel('Process Temp (K)')
plt.savefig('static/4_thermal_density.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# --- الرسم الخامس: مصفوفة الارتباط الكاملة (Heatmap) ---
# الخلاصة لكل العوامل الميكانيكية في رسمة واحدة
plt.figure(figsize=(12, 10))
numeric_cols = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_cols.corr(), annot=True, cmap='RdBu_r', fmt='.2f', linewidths=0.5)
plt.title('Final Correlation Matrix: Predictive Maintenance Factors')
plt.savefig('static/5_correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
```
